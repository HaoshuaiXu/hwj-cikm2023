```python
import pandas as pd

wiki80_training_data = pd.read_json("./dataset/wiki80/wiki80_train.txt", lines=True)

wiki80_training_data
```

```python
type(wiki80_training_data['token'][0])
```

```python
from transformers import RobertaTokenizer

tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
```

```python
tokenizer.encode_plus(wiki80_training_data['token'][0])
```

```python
import json

with open("./dataset/wiki80/wiki80_rel2id.json") as f:
    relation2id = json.load(f)

relation2id
```

```python
wiki80_training_data_copy = wiki80_training_data.copy()

for k, v in relation2id.items():
    wiki80_training_data_copy.replace(to_replace=k, value=v, inplace=True)
```

```python
wiki80_training_data_copy
```

```python
wiki80_training_data_copy.rename(
    columns={'token': 'text',
     'relation': 'label'},
    inplace=True
)

wiki80_training_data_copy[wiki80_training_data_copy['label'] < 42]
```
