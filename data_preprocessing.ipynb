```python
with open("./dataset/TEST_FILE_CLEAN.TXT") as f:
    test_sents = [i.strip().split('\t') for i in f.readlines()]

test_sents[0]
```

```python
test_sent_ids = []
test_sent_texts = []

for ts in test_sents:
    test_sent_ids.append(ts[0])
    test_sent_texts.append(ts[1])

print(test_sent_ids[0])
print(test_sent_texts[0])
```

```python
print(len(test_sent_ids)==len(test_sent_texts))
```

```python
with open("./dataset/TEST_FILE_KEY.TXT") as f:
    test_labels = f.readlines()
    test_labels = [tl.strip().split('\t') for tl in test_labels]

test_labels
```

```python
test_labels_labels = []

for tl in test_labels:
    test_labels_labels.append(tl[1])

print(test_labels_labels[0])
print(len(test_labels_labels))
```

```python
with open("./dataset/semeval_test.tsv", 'a') as f:
    for i in range(0, len(test_sent_ids)):
        f.writelines("%s\t%s\t%s\n" % (test_sent_ids[i], test_sent_texts[i], test_labels_labels[i]))
```
