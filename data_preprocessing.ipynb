```python
with open("./dataset/legacy/TEST_FILE_FULL.TXT") as f:
    test_sents = [i.strip().split('\t') for i in f.readlines()]

test_sents
```

```python
with open("./dataset/semeval_testset_double.tsv", '+a') as f:
    for i in range(0, len(test_sents), 4):
        idx = test_sents[i][0]
        text = test_sents[i][1]
        label = test_sents[i+1][0]
        comment = test_sents[i+2][0]
        f.writelines(f"{idx}\t{text}\t{label}\t{comment}\n")
```

```python
import pandas as pd


testset = pd.read_csv("./dataset/semeval_testset_double.tsv", delimiter='\t')
testset.head()
```

```python
with open("./dataset/semeval_label_double.tsv") as f:
    labelset = [l.strip() for l in f.readlines()]
labelset
```

```python
for i, v in enumerate(labelset):
    testset.replace(to_replace=v, value=i, inplace=True)

testset.head()
```

```python
testset.to_csv("./dataset/semeval_testset_double_classcode.tsv", sep='\t', index=None)
```
